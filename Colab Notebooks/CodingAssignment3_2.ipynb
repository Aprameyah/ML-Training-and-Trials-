### Importing the required liberaries


```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import PolynomialFeatures
```

### Importing the dataset 

```
dataset = pd.read_csv("/content/sample_data/sat-gpa-dataset.csv")
X = dataset.iloc[:,0].values
y = dataset.iloc[:,1].values
X = np.reshape(X,(-1, 1))
y = np.reshape(y,(-1, 1))
```

### Visualizing the Data 

```
plt.scatter(X, y, color = "blue", label = "Datapoints")
plt.title("GPA Vs SAT Dataset")
plt.xlabel("SAT Score")
plt.ylabel("GPA")
plt.legend()
plt.show()
```

### Performing the train test split

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

### Creating a Polynomial Regression Function
[Reference to fix plotting problem](https://stackoverflow.com/questions/37414916/pythons-matplotlib-plotting-in-wrong-order)



```
def PolynomialRegression(degree):

  poly_model = PolynomialFeatures(degree)
  X_poly_train = poly_model.fit_transform(X_train)
  X_poly_test = poly_model.fit_transform(X_test)
  poly_reg = LinearRegression()
  poly_reg.fit(X_poly_train, y_train)
  y_poly_pred = poly_reg.predict(X_poly_test)
  y_polytrain_pred = poly_reg.predict(X_poly_train)

  # Visualizing the data 
  plt.figure()
  plt.scatter( X_train, y_train, color = "red" ,label = "training set")
  plt.scatter( X_test, y_test, color = "magenta" ,label = "test set")
  plt.scatter( X_test, y_poly_pred, color = "blue" ,label = "predicted set")
  x_temp, y_temp = zip(*sorted(zip(X_test, y_poly_pred)))
  plt.plot( x_temp, y_temp, color = "black" ,label = "Model function")
  plt.title("GPA Vs SAT Dataset")
  plt.xlabel("SAT Score")
  plt.ylabel("GPA")
  plt.legend()
  plt.show()

  #Getting the error values
  #print('MSE for degree',degree,"is",mean_squared_error(y_test, y_poly_pred))
  #print('MAE for degree',degree,"is",mean_absolute_error(y_test, y_poly_pred))

  print('MSE for testing dataset for degree',degree,"is",mean_squared_error(y_test, y_poly_pred, squared = True))
  print('MSE for training dataset for degree',degree,"is",mean_squared_error(y_train, y_polytrain_pred, squared = True))
```
