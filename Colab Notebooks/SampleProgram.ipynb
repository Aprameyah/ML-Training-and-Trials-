# Lists

```
tools = ['pen', 'hammer', 'lever']
print(tools)

tools_slice = tools[2:3] # ['hammer', 'lever']
print(tools_slice)

tools_slice[0] = 'nail'
print(tools_slice)

# Original list is unaltered:
print(tools) # ['pen', 'hammer', 'lever']
```

# For loop

```
dog_breeds = ["boxer", "bulldog", "shiba inu"]

# Print each breed:
for breed in dog_breeds:
  print(breed)
```
