### Importing the required liberaries




```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

### Importing the dataset 


```
dataset = pd.read_csv("/content/sample_data/sat-gpa-dataset.csv")
X = dataset.iloc[:,0].values
Y = dataset.iloc[:,1].values
```

### Checking the values  

```
X = np.reshape(X,(-1, 1))
Y = np.reshape(Y,(-1, 1))
```

### Visualizing the data

```
plt.scatter(X, Y, color = "blue", label = "Dataset")
plt.title("GPA Vs SAT Dataset")
plt.xlabel("SAT Score")
plt.ylabel("GPA")
plt.legend()
plt.show()
```

### Creating a linear regression model and visualizing it 

```
regressor_1 = LinearRegression()
regressor_1.fit(X, Y)

# Visualizing the data
plt.scatter(X, Y, color = "red", label = "dataset")
plt.plot(X, regressor_1.predict(X), color = "blue", label ="model function")
plt.title("GPA Vs SAT Dataset")
plt.xlabel("SAT Score")
plt.ylabel("GPA")
plt.legend()
plt.show()
```

### Train test split

```
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state = 0)
```

```
plt.scatter(X_train, Y_train, color = "red", label = "training set")
plt.scatter(X_test, Y_test, color = "blue", label = "test set")
plt.title("GPA Vs SAT Dataset")
plt.xlabel("SAT Score")
plt.ylabel("GPA")
plt.legend()
plt.show()
```

### Training using train_split & Visualizing

```
regressor_2 = LinearRegression()
regressor_2.fit(X_train, Y_train)
Y_predict = regressor_2.predict(X_test)

# Visualizing the data

plt.scatter(X_test, Y_test, color = "red", label = "test set")
plt.scatter(X_train, Y_train, color = "magenta", label = "training set")
plt.scatter(X_test, Y_predict, color = "blue", label = "predicted values")
plt.plot(X_test, Y_predict, color = "green", label = "model function")
plt.title("GPA Vs SAT Dataset")
plt.xlabel("SAT Score")
plt.ylabel("GPA")
plt.legend()
plt.show()
```

### Evaluting the performance 

```
print("MSE", mean_squared_error(Y_test, Y_predict))
print("MAE", mean_absolute_error(Y_test, Y_predict))
```
